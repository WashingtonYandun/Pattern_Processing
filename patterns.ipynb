```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## Data Loading and Preprocessing

-   We will load the data and see the first few rows of the data.


```python
mnist_train_data = pd.read_csv("train.csv")
mnist_train_data.head()
```

The data is loaded using the `read_csv` function from the `pandas` library. The `head` function is used to see the first few rows of the data.
The data is composed of 785 columns. The first column is the label and the rest of the columns are the pixel values of the image.
In the next step, we will see the distribution of the labels.


```python
for i in range(0, 9):
    plt.subplot(330 + 1 + i)
    plt.imshow(mnist_train_data.iloc[i*2, 1:].values.reshape(28, 28), cmap=plt.get_cmap('gray'))
```

```python

dex = np.array(mnist_train_data.iloc[3, 1:])
plt.imshow(dex.reshape(28,28), cmap='Greys')
dex.max(), dex.std(), dex.mean(), dex.min()

plt.figure(figsize=(8,8))
sns.heatmap(dex.reshape(28,28), square=True, cbar=False, annot=True, fmt='g', annot_kws={'fontsize': 6})
```
